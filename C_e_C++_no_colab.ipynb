Usando o comando switch, implemente um algoritmo que leia um inteiro entre 1 e 7 e imprima o dia da semana correspondente a esse número. Isto é, Domingo!, se 1, Segunda-feira!, se 2, e assim por diante. Além disso, caso o usuário insira qualquer número diferente de 1-7, a mensagem "Numero invalido!" deve ser impressa.

Por exemplo, o seu algoritmo deve ser capaz de, usando a seguinte ENTRADA:

Domingo!

No caso da ENTRADA:

9

```python
#include <stdio.h>
#include <math.h>
int main(){

	int s;
	scanf("%d",&s);

	switch(s){
		case 1:
			printf("Domingo!\n");
			break;
		case 2:
			printf("Segunda-feira!\n");
			break;
		case 3:
			printf("Terca-feira!\n");
			break;
		case 4:
			printf("Quarta-feira!\n");
			break;
		case 5:
			printf("Quinta-feira!\n");
			break;
		case 6:
			printf("Sexta-feira!\n");
			break;
		case 7:
			printf("Sabado!\n");
			break;
		default:
			printf("Numero Invalido!");
			break;

	}


}
```

Implemente o Algoritmo de Euclides para encontrar o MDC (Máximo Divisor Comum).

A ideia principal no Algoritmo de Euclides é que o MDC pode ser calculado recursivamente, usando o resto da divisão como entrada para o próximo passo, o que é baseado na seguinte propriedade do MDC:

 MDC ( a , b ) = MDC ( b , r )

onde r é o resto da divisão de a com b.

[clique aqui para mais detalhes do algoritmo de Euclides]

Implemente duas versões do algoritmo:
Versão recursiva;
Versão iterativa (sem recursão).
Considere:
Os argumentos de entrada e o retorno são do tipo long.
Existem dois argumentos de entrada, e o primeiro é maior que o segundo.
O retorno da função é o MDC dos dois valores de entrada.
Entrada do programa, em ordem:
NÚMERO INTEIRO com o primeiro valor (maior);
NÚMERO INTEIRO com o segundo valor (menor);
Saída do programa:
"Profundidade da Recursao = X e MDC = Y", onde:
X é a profundidade da recursão da função recursiva (número máximo de chamadas simultâneas);
Y é o MDC.
Observações:
Implemente apenas as funções euclides_recursivo e euclides_iterativo, conforme o arquivo inicial fornecido;
Não imprima nada, ou seja, não utilize printf, puts, etc;
O programa retornará uma mensagem de erro se a função euclides_iterativo for recursiva.

```python
#include "euclides.h"

long euclides_recursivo(long m, long n)
{

    if (n==0) return m;
    return euclides_recursivo (n, m % n);
}

long euclides_iterativo(long m, long n)
{
    long r;

     do
       {
         r = m % n;
         m = n;
         n = r;
       }
     while (r != 0);

     return m;
}
```

```python
#include <stdio.h>
#include <stdlib.h>
#include<math.h>

struct Dis{
	int x1;
	int y1;
	int x2;
	int y2;

};


int main(){

	struct Dis D;
	scanf("%d",&D.x1);
	scanf("%d",&D.y1);
	scanf("%d",&D.x2);
	scanf("%d",&D.y2);

	float distancia = sqrt((D.x2 - D.x1) *(D.x2 - D.x1) + (D.y2 - D.y1) * (D.y2 - D.y1) );


	printf("%.1f",distancia);

}
```
